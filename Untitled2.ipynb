Given a sentence, identify all unique words and display:


```python

sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)
print("Count of unique words:", len(unique_words))
print("Unique words:", unique_words)
```

Store employee names and their salaries, and determine which employee has the highest salary.


```python

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_employee = max(employees, key=employees.get)
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

Find the maximum and minimum values from a list of numbers.


```python

numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num


print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

 Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]

present = 0
for day in attendance:
    if day == "P":
        present += 1

percentage = (present / len(attendance)) * 100
print("Attendance Percentage:", percentage)
```

Remove Duplicate Phone Numbers


```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

Count Character Frequency


```python
text = "pythonp"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```
